***************************************** WIN RATE **********************************************
For each champion, we're looking for the items and spells with the most win rate. If somehow the item is repeated, we select the following one.

```python
position = -1

for id in final_dataframe["id"]:

    print("This is for champ with id: "+str(id))

    position+=1

    for i in range(6):
        num = 0
        item = (players_df[(players_df['championId']== id ) & (players_df['stats.win']== True)]['stats.item'+str(i)].value_counts()).index[num]
        
        j = i-1
        llista =[]
        
        #Creem una llista amb els elements anteriors.

        while j>=0:

            llista.append(final_dataframe.at[position,"item"+str(j)])
            j-=1

        #Mirem que l'element afegit no sigui igual a 0 o ja estigui implementat

        while item in llista or int(item)==0:
            num +=1
            item = (players_df[(players_df['championId']== id ) & (players_df['stats.win']== True)]['stats.item'+str(i)].value_counts()).index[num]
        

        print("item"+str(i)+": "+str(int(item)))

        final_dataframe.at[position,"item"+str(i)] = int(item)
```
