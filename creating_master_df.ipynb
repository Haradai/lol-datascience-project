Import all dataframes from the pickle files provided in kaggle

```python
import pandas as pd
```

```python
match_df = pd.read_pickle("match_data_version1.pickle")
```

```python
win_df = pd.read_pickle("match_winner_data_version1.pickle")
```

```python
los_df = pd.read_pickle("match_loser_data_version1.pickle")
```

We delete columns that we don't want

```python
a = win_df.drop("win",axis=1)
a.columns
```

```python
#we change column names to differenciate winner and looser team data 
a.columns = ['teamId_w', 'firstBlood_w', 'firstTower_w', 'firstInhibitor_w', 'firstBaron_w',
       'firstDragon_w', 'firstRiftHerald_w', 'towerKills_w', 'inhibitorKills_w',
       'baronKills_w', 'dragonKills_w', 'vilemawKills_w', 'riftHeraldKills_w',
       'dominionVictoryScore_w', 'bans_w', 'gameId']
```

```python
aux = pd.merge(a,match_df,on="gameId",how="left")
```

```python
#delete not needed columns
aux = aux.drop("gameCreation",axis=1)
aux = aux.drop("gameType",axis=1)
aux = aux.drop("platformId",axis=1)
aux = aux.drop("queueId",axis=1)
aux = aux.drop("status.message",axis=1)
aux = aux.drop("status.status_code",axis=1)
```

```python
aux.columns
```

```python
b = los_df.drop("win",axis=1)
b.columns
```

```python
b.columns = ['teamId_l', 'firstBlood_l', 'firstTower_l', 'firstInhibitor_l', 'firstBaron_l',
       'firstDragon_l', 'firstRiftHerald_l', 'towerKills_l', 'inhibitorKills_l',
       'baronKills_l', 'dragonKills_l', 'vilemawKills_l', 'riftHeraldKills_l',
       'dominionVictoryScore_l', 'bans_l', 'gameId']
```

```python
master_df =  pd.merge(aux,b,on="gameId",how="left")
```
